```python
target = pd.read_csv(r'C:\Users\Khangjrakpam Arjun\OneDrive - McKinsey & Company\Documents\ARJUN\ML\COMPETITIONS\Amex\amex-default-prediction\train_labels.csv')
train = pd.merge(
    target,
    train,
    how="inner",
    on=None,
    left_on='customer_ID',
    right_on='customer_ID',
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
```

```python
for i in train.columns:
    print(i)
```

```python
#features = train.drop(['customer_ID', 'S_2', 'target'], axis=1).columns.to_list()
features = train.drop(['customer_ID',  'target'], axis=1).columns.to_list()
cat_features = [
        'D_48','B_9',
         'D_58',
         'D_44',
         'B_7',
         'B_3',
         'B_4',
         'B_1', 'B_22','R_1',
         'B_8',
         'S_3',
         'D_61',
         'R_3',
         'D_70',
         'D_66',
         'D_68',
         'B_30',
         'B_31',
         'B_38',
         'D_114',
         'D_116',
         'D_117',
         'D_120',
         'D_126',
         'D_63',
         'D_64',
]

num_features = [col for col in features if col not in cat_features]
```

```python
train_num_agg = train.groupby("customer_ID")[num_features].agg(['mean', 'std', 'min', 'max', 'last','median'])
train_num_agg.columns = ['_'.join(x) for x in train_num_agg.columns]
train_cat_agg = train.groupby("customer_ID")[cat_features].agg(['count', 'last', 'nunique'])
train_cat_agg.columns = ['_'.join(x) for x in train_cat_agg.columns]
train_target = (train.groupby("customer_ID").tail(1).set_index('customer_ID', drop=True).sort_index()["target"])
train = pd.concat([train_num_agg, train_cat_agg, train_target], axis=1)
train.reset_index(inplace = True)
train.to_pickle("../preprocessed_data/train_agg_imputed_with_median.pkl")
```

```python
test_num_agg = test.groupby("customer_ID")[num_features].agg(['mean', 'std', 'min', 'max', 'last','median'])
test_num_agg.columns = ['_'.join(x) for x in test_num_agg.columns]
test_cat_agg = test.groupby("customer_ID")[cat_features].agg(['count', 'last', 'nunique'])
test_cat_agg.columns = ['_'.join(x) for x in test_cat_agg.columns]
test = pd.concat([test_num_agg, test_cat_agg], axis=1)
test.reset_index(inplace = True)

test.to_pickle("../preprocessed_data/test_agg_imputed_with_median.pkl")
```
